Определяем по какой из наибольшей карреляции (свыше 0,6) строить зависимость

```python
df['date'] = pd.to_datetime(df['date'])
df['numeric_date'] = df['date'].astype('int64')
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(9, 8))
corr_matrix = numeric_df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
upp_mat = np.triu(corr_matrix)

sns.heatmap(corr_matrix, annot=True, linewidths=.5,
            cmap='coolwarm', mask=upp_mat)
plt.show()
```

Комментарии:
- распределение цены в зависимости от года постройки имеет U-образную форму;
- цена на дома, построенные с 1900-х до 1960-х, постепенно опускалась;
- при этом цена на дома, построенные в 1940-х годах, принимает низкие значения;
- цена на дома, построенные с 1960-х годов, вновь начала расти;
- для домов, построенных в 2020-х годах, цена почти такая же, как и для домов, построенных в 1900-х годах;
- могу предположить, что старые дома, которые стоят больше 1 века, имеют особую историческую значимость и поэтому 
  их цена находится на одном уровне с ценами домов, построенных в 2000-х годах.

```python
plt.figure(figsize=(8, 5))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.0)

plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

Комментарии:
- с увеличением уровня класса (grade - качество конструкции и дизайна дома) растет разбег по цене;
- для домов с 7-11 классом наблюдается максимальное количество выбросов.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])

plt.title('Распределение цен по береговой линии')
plt.xlabel('Цена')
plt.ylabel('Наличие набережной');
```

Комментарии:
- уровень цен 50% домов, не выходящих набережную, значительно ниже, чем цена 50% домов, выходящих на нее ;
- для домов без выхода к набережной выбросов намного больше (видимо заинтересованность риелторов низкая по получаемой 
выручке и поэтому качество обработки данных теряеться);
- разбег цены для домов с выходом к набережной больше, чем для домов без него.

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Стоимость/Этажность дома');
```

Комментарии:

- минимальный разбег по ценам  для домов в 3,5 этажей;

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = df['price'], y = df['condition'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Состояние дома')
plt.title('Стоимость/Оценка состояния');
```
