```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene

# Данные
football_players = [173, 175, 180, 178, 177, 185, 183, 182]
hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]
weightlifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]

# Создаем DataFrame для удобства анализа
import pandas as pd
data = {
    'Рост': football_players + hockey_players + weightlifters,
    'Вид спорта': ['Футбол']*len(football_players) +
                  ['Хоккей']*len(hockey_players) +
                  ['Штанга']*len(weightlifters)
}
df = pd.DataFrame(data)

# 1. Описательная статистика
print("Описательная статистика:")
print(df.groupby('Вид спорта')['Рост'].describe())
print("\n")

# 2. Проверка на нормальность распределения (тест Шапиро-Уилка)
print("Тест на нормальность распределения:")
for sport in ['Футбол', 'Хоккей', 'Штанга']:
    stat, p = shapiro(df[df['Вид спорта'] == sport]['Рост'])
    print(f"{sport}: p-значение = {p:.4f}")
print("\n")

# 3. Проверка гомогенности дисперсий (тест Левена)
stat, p = levene(football_players, hockey_players, weightlifters)
print(f"Тест Левена на гомогенность дисперсий: p-значение = {p:.4f}\n")

# 4. Однофакторный дисперсионный анализ (ANOVA)
f_statistic, p_value = stats.f_oneway(football_players, hockey_players, weightlifters)
print(f"Результаты ANOVA:")
print(f"F-статистика: {f_statistic:.4f}")
print(f"p-значение: {p_value:.4f}\n")

# 5. Визуализация
plt.figure(figsize=(10, 6))
sns.boxplot(x='Вид спорта', y='Рост', data=df)
plt.title('Распределение роста спортсменов по видам спорта')
plt.show()

# Интерпретация результатов
alpha = 0.05
print("Выводы:")
if p_value < alpha:
    print("Есть статистически значимые различия в среднем росте между группами.")
else:
    print("Нет статистически значимых различий в среднем росте между группами.")
```
